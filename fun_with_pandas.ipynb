## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
```

```python
rawdata
```

```python
physical_df = rawdata[0]
```

```python
physical_df
```

```python
digital_df = rawdata[3]
```

```python
digital_df
```

```python
merged_df = pd.concat([physical_df, digital_df])
list(merged_df)
merged_df
```

```python
merged_df["Digital or physical?"] = "who knows" #creating column for later for loop
merged_df['Sales (in millions)'] = merged_df['Sales (in millions)'].replace(r'[^\d.]+', '', regex=True) #removing letters
merged_df['Sales (in millions)'] = pd.to_numeric(merged_df['Sales (in millions)'], errors='coerce') #making sure values are numeric
merged_df = merged_df.sort_values(by="Sales (in millions)", ascending=False) #sorting by highest value in sales column
merged_df = merged_df.reset_index(drop=True)
```

```python

for idx, df_row in merged_df.iterrows():
    if df_row["Single"] in digital_df["Single"].values:
        merged_df.at[idx, 'Digital or physical?'] = 'digital'
    elif df_row['Single'] in physical_df['Single'].values:
        merged_df.at[idx, 'Digital or physical?'] = 'physical'

merged_df
```

```python
print(merged_df.info())
print(merged_df['Released'].unique())
birth_year = 2000
merged_df['Released'] = merged_df['Released'].astype(int)
exact_match = merged_df[merged_df['Released'] == birth_year]
exact_match #there is no exact match
```

```python
#since there's no song for my birth year, i create a loop
merged_df['year_diff'] = (merged_df['Released'] - birth_year).abs()
merged_df
closest_year = merged_df.sort_values(by='year_diff').head(1)
closest_year

```

```python
if not exact_match.empty:
    print(exact_match)

else:
    print(closest_year)
```

```python
fifteen_year = 2015
fifteen_song = merged_df[merged_df['Released'] == fifteen_year]
fifteen_song
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
#found a way better way to do this than i used for the first task
combined_df = pd.concat(rawdata[:6], ignore_index=True)
combined_df
```

```python
#extracting the year by requesitng four digits 
combined_df['Year'] = combined_df['Launch date'].str.extract(r'(\d{4})')
#counting launches per year by counting how many times each year appears in the launch data column
launches_per_year = combined_df['Year'].value_counts()
#finding the year by looking at year with highest count
most_launches_year = launches_per_year.idxmax()
most_launches_year
```

```python
#using same tempalte but for month instead
combined_df['Month'] = combined_df['Launch date'].str.extract(r'(January|February|March|April|May|June|July|August|September|October|November|December)', expand=False)
launches_per_month = combined_df['Month'].value_counts()
most_launches_month = launches_per_month.idxmax()
most_launches_month
```

```python
months_sorted = launches_per_month.sort_values(ascending=False)
months_sorted
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
combined_df2 = pd.concat(rawdata[3:11], ignore_index=True)
combined_df2
```

```python
#writing code that extracts the decade of each row and puts it into a decade column
combined_df2['Year Debuted'] = pd.to_numeric(combined_df2['Year Debuted'], errors='coerce') #make the values numeric
combined_df2['Decade'] = (combined_df2['Year Debuted'] // 10) * 10 #this removes the last digit from the year, and adds a zero to define the decade
villains_per_decade = combined_df2['Decade'].value_counts()
villains_per_decade
```

```python
villains_per_company = combined_df2['Company'].value_counts() #counting number of appearances
company_ranking_df = villains_per_company.reset_index() #converts results into a dataframe with numbered index
company_ranking_df.columns = ['Company', 'Number of Supervillains'] #renaming columns
company_ranking_df = company_ranking_df.sort_values(by='Number of Supervillains', ascending=False)#most to least
company_ranking_df
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
#list(df)
```

```python

# add id column
df['Participant_id'] = df.index 

# shaping column 
df2 = pd.melt(df,
                  id_vars=['Participant_id', 'Year'],  # columns from old df
                  value_vars=['Reading_NoInt', 'Naming_Int', 'Naming_NoInt', 'Reading_Int'],  # get task info in new column
                  var_name='Task',  
                  value_name='RT') 

print(df2)
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
new_df = df2.groupby('Task')['RT'].mean().reset_index()

# rename column
new_df.rename(columns={'RT': 'Mean_RT'}, inplace=True)
print(new_df)
```
